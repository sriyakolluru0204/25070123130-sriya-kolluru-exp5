```python

'''3. A dictionary stores usernames and passwords.
Write a Python program to validate user login.'''
users = {"admin": '1234', "student": "abcd"}
user_name = input("Username: ")
pw = input("Password: ")

if users.get(user_name) == pw:
  print("Login Successfull")
else:
  print("Invalid Username or Password")
```
